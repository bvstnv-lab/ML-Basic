## Задача 4

```python
import numpy as np

# задаем исходные векторы
a1=np.array([2,-5])
a2=np.array([-1,3])
x=np.array([1,-4])

# проверяем линейную зависимость а1 и а2
print(a1[0]/a2[0] == a1[1]/a2[1])

# создаем новый базис
B=np.array([a1,a2]).T
print(B)
```

a) Координаты вектора в старом и новом базисе связаны следующим образом:

$S\cdot[X]_S = B\cdot[X]_B$

Координаты вектора в новом базисе находятся по выражению:

$[X]_B = B^{-1}\cdot[X]_S$

```python
xb=np.linalg.inv(B).dot(x)
print(xb)
```

Получили новые координаты:
    
$[X]_B = \begin{pmatrix} -1 \\ -3 \end{pmatrix}$

b) Координаты вектора в стандартном базисе определяются по выражению:


$[Y]_S = B\cdot[Y]_B$

```python
yb=np.array([1,1])
print(B.dot(yb))
```

Получили координаты вектора y в стандартном базисе:
    
$[Y]_S = \begin{pmatrix} 1 \\ -2 \end{pmatrix}$

## Задача 5

```python
import numpy as np
import matplotlib.pyplot as plt

# генерируем случайную матрицу 100х100 с элементами в диапазоне [-1,1)
n=100
np.random.seed(0)
A=2*np.random.rand(n,n)-1

# выполняем SVD-разложение
U,S,Vh=np.linalg.svd(A)

# восстанавливаем матрицу с приближением и считаем ошибку
E=[]
for r in range(2, n+1):
    A_reborn=np.dot(U[:,:r]*S[:r], Vh[:r,:])
    E.append(float((sum(sum((A-A_reborn)**2))**0.5)))

# смотрим график и радуемся)
plt.plot([r for r in range(2, n+1)], E)
plt.show()
```

# Задача 4

Первый дифференциал приведенной функции вычисляется по формуле:
$$
df(x)=f'(x) \cdot dx=\left( \frac{1}{3} \cdot {||x||}_2^3 \right)' \cdot  \left( {||x||}_2 \right)' \cdot dx
$$

Евклидова норма представляет собой функцию вида:
$${||x||}_2=\sqrt{\sum\limits_{i=1}^{n} x_i^2}$$


Ее частные производные можно определить по формуле сложной производной:
$$\frac{\partial f}{\partial x_i}=\left( \sqrt{\sum\limits_{i=1}^{n} x_i^2} \right)' \cdot \left( \sum\limits_{i=1}^{n} x_i^2 \right)' = \frac{1}{2} \cdot \frac{1}{\sqrt{\sum\limits_{i=1}^{n} x_i^2}} \cdot 2 \cdot x_i = \frac {x_i}{||x||_2}$$

Получим дифференциал первоначальной функции по каждой координате:
$$
df(x_i)={||x||}_2^2 \cdot \frac {x_i}{||x||_2} \cdot dx_i = {||x||}_2 \cdot x_i \cdot dx_i
$$

Или иначе:
$$
df(x)=<{||x||}_2 \cdot x_i, \ dx>
$$

Где градиент имеет вид:
$$
\nabla f(x) = {||x||}_2 \cdot x_i
$$

# Задача 6

Имеем:
$$
df(x)=f'(x) \cdot dx=\left( {||A \cdot x||}_2 \right)'\cdot dx
$$

Евклидова норма произведения матрицы $m$ x $n$ на вектор $n$:
$${||A \cdot x||}_2=\sqrt{\sum\limits_{i=1}^{m} \left( \sum\limits_{j=1}^{n} A_{ij} \cdot x_j \right)^2}$$


Ее частные производные можно определить по формуле сложной производной:
$$\frac{\partial f}{\partial x_i}=\left( \sqrt{\sum\limits_{i=1}^{m} \left( \sum\limits_{j=1}^{n} A_{ij} \cdot x_j \right)^2} \right)' \cdot \left( \sum\limits_{i=1}^{m} \left( \sum\limits_{j=1}^{n} A_{ij} \cdot x_j \right)^2 \right)' = \frac{1}{2} \cdot \frac{1}{{||A \cdot x||}_2} \cdot 2 \cdot \sum\limits_{i=1}^{m} \left( A_{ij} \right)^2 \cdot x_j$$

Получим:
$$\frac{\partial f}{\partial x_i}=\frac{1}{{||A \cdot x||}_2} \cdot \sum\limits_{i=1}^{m} \left( A_{ij} \right)^2 \cdot x_j$$

Дифференциал функции:
$$
df(x)=<\frac{1}{{||A \cdot x||}_2} \cdot \sum\limits_{i=1}^{m} \left( A_{ij} \right)^2 \cdot x_j, \ dx>
$$

Где градиент имеет вид:
$$
\nabla f(x) = \frac{1}{{||A \cdot x||}_2} \cdot \sum\limits_{i=1}^{m} \left( A_{ij} \right)^2 \cdot x_j
$$
