## Задание 1: Получение однозначного числа

Пользователь вводит целое число, программа складывает все цифры числа, с полученным числом — то же самое, и так до тех пор, пока не получится однозначное число.

<br>

**Примеры:**
* `545   -> 5`
* `12345 -> 6`

````python
while True:
    ''' повторяем ввод, пока не введено целое положительное число, либо 0
        также отбрасываем числа с ведущим 0 '''
    num = input('Введите целое положительное число ')
    if num.isnumeric() and (len(num) == 1 or int(num[0])):
        break

while int(num) >= 10:
    ''' повторяем операции, пока число не станет однозначным '''
    num_new = 0

    for i in range(len(num)):
        ''' получаем сумму цифр числа '''
        num_new += int(num[i])
    num = str(num_new)

print(num)
````

## Задание 2: Кинотеатр

Дан список списков, каждый вложенный список состоит из 1 и 0, количество вложенных списков — количество рядов. Пользователь вводит, сколько билетов ему требуется. Программа должна найти ряд, где можно приобрести нужное количество билетов (места должны быть рядом). Если таких рядов несколько, то ближайший к экрану (ближайшим считается нулевой ряд). Если таких мест нет, то вывести False.

<br>

**Примеры:**
* `[[0,1,1,0], [1,0,0,0], [0,1,0,0]], 2 -> 1`
* `[[0,1,1,0], [1,0,1,0], [1,1,0,1]], 2 -> False`

````python
places = [[0,1,1,0], [1,0,1,0], [1,1,0,1]] 
''' по условию считаем список заданным '''  

while True:
    ''' повторяем ввод, пока не введено целое положительное число
        также отбрасываем 0 и числа с ведущим 0, число не более 4 (мест в ряду) '''
    tickets = input('Введите количество билетов ')
    if tickets.isnumeric() and int(tickets[0]) and int(tickets) < 5:
        break

tick_str = int(tickets)*'0'

flag = 0
for line, coll in enumerate(places):
    ''' в цикле преобразуем ряд мест в строку для дальнейшего сравнения'''
    coll_str = ''.join(map(str, coll))
    if tick_str in coll_str:
        ''' находим желаемую последовательность - печатаем, поднимаем флаг и выходим'''
        print(line)
        flag = 1
        break
if not flag:
    ''' если мест не нашли, то флаг не поднят'''
    print('False')
````

## Задание 3: Алгоритм RLE

Необходимо написать упрощенную версию алгоритма RLE. 
Алгоритм RLE объединяет подряд идущие символы в коэффициент и символ.

<br>

**Примеры:**
* `asssdddsssddd -> 1a3s3d3s3d`
* `aaabbbbccccc  -> 3a4b5c`
* `abcba         -> 1a1b1c1b1a`

````python
line = str(input())
c=0         # вводим счетчик символов
a=line[0]   # стартуем с первого символа
for i in range(len(line)):
#считаем одинаковые символы, если дошли до конца строки - выводим итог
    if line[i] == a:
        c += 1
        if i==len(line)-1:
            print(c,a, sep='', end='')

#при смене символа выводим предыдущий итог и начинаем сначала, если это последний символ - выводим итог
    else:
        print(c,a, sep='', end='')
        a = line[i]
        c = 1
        if i==len(line)-1:
            print(c,a, sep='', end='')
````

## Задание 4: Шифр Цезаря

Реализуйте программу для шифрования текста с помощью шифра Цезаря.  
Шифр Цезаря — это метод шифрования, при котором каждая буква в тексте заменяется буквой, стоящей на фиксированное число позиций дальше в алфавите.

### Требования
1. Программа должна принимать на вход:
   - Строку текста для шифрования
   - Ключ шифра (число от 0 до 25)

2. Правила шифрования:
   - Шифруются только латинские буквы (a-z, A-Z)
   - Сохраняется регистр букв (заглавные остаются заглавными)
   - Пробелы, цифры и знаки препинания остаются без изменений
   - Сдвиг циклический (после 'z' идет 'a', после 'Z' идет 'A')

### Примеры

* `"Hello World", 3      -> "Khoor Zruog"`
* `"Python is great!", 5 -> "Udymts nx lwjfy!"`
* `"XYZ abc", 3          -> "ABC def"`
* `"Test 123", 1         -> "Uftu 123"`

### Алгоритм
* Для каждого символа в строке:
   - Если это латинская буква, найти её позицию в алфавите
   - Добавить к позиции значение ключа
   - Если результат больше 25, использовать остаток от деления на 26
   - Преобразовать обратно в букву, сохранив регистр
   - Если это не буква, оставить символ без изменений


````python
lat_s = 'abcdefghijklmnopqrstuvwxyz'
lat_l = lat_s.upper()
pos = 0

line=input('Введите латинский текст ')
while True:
    ''' повторяем ввод, пока не введено целое положительное число от 0 до 25
        отбрасываем числа с ведущим 0 '''
    key = input('Введите ключ ')
    if key.isnumeric() and int(key[0]) and int(key) < 26  or key =='0':
        break

for i in range(len(line)):
    if line[i] in lat_s:
        # находим новую позицию строчной буквы"
        pos = (lat_s.rfind(line[i]) + int(key)) % 26
        print(lat_s[pos], end='')
    elif line[i] in lat_l:
        # находим новую позицию прописной буквы
        pos = (lat_l.rfind(line[i]) + int(key)) % 26
        print(lat_l[pos], end='')
    else:
        # иначе не изменяем символ
        print(line[i], end='')
````

## Задание 5: Табель успеваемости

Пользователь в бесконечном цикле (пока не будет введена пустая строка) вводит строки вида: 'название предмета' 'фамилия ученика' 'оценка'. После окончания ввода программа выводит в консоль название предмета, далее список учеников и все их оценки в виде таблицы.

<br>

**Примеры:**

<br>

*Ввод:*
```python
Математика Иванов  5
Математика Иванов  4
Литература Иванов  3
Математика Петров  5
Литература Сидоров 3
Литература Петров  5
Литература Иванов  4
Математика Сидоров 3
Математика Петров  5
```

<br>

*Результат:*
```python
Математика # вводим название предмета

# выводим список учеников и их оценки
Иванов  5 4
Петров  5 5
Сидоров 3
```

```python
Литература # вводим название предмета

# выводим список учеников и их оценки
Иванов  3 4
Сидоров 3
Петров  5
```

````python
tab1 = {}
tab2 = {}
while True: #повторяем ввод до пустой строки
    line=input()
    if line=='':
        break
    else: #строку делим на части, в словаре tab1 ключи - предмет+фамилия, значения - оценка
        if tab1.get(line.split()[0] +' '+ line.split()[1]) == None: #костыль в виде 0 для нового ключа, чтобы считалось дальше
            tab1[line.split()[0] +' '+ line.split()[1]] = '0'
        #заполняем tab1 по ключу:уже имеющееся значение (None заменен на 0)+оценка
        #имеем в виду, что если ключ уже попадался в предыдущей итерации, то к значению (оценке) добавляем новую оценку
        tab1[line.split()[0]+' '+line.split()[1]] = str(tab1.get(line.split()[0]+' '+line.split()[1]))+(line.split()[2])+' '

for i in tab1:
    tab1[i]=(tab1.get(i)).removeprefix('0') #убираем 0, которые сами же и добавили
    if tab2.get(i.split()[0]) == None:
        tab2[i.split()[0]] = '0' #тут аналогично описанному выше
    #тут мы хотим создать словарь с ключом - предмет, значением - фамилия+оценка,оценка..,фамилия+оценка,оценка...
    tab2[i.split()[0]] = (str(tab2.get(i.split()[0])) + ' ' + str(i.split()[1])+ ' ' + tab1.get(i)).removeprefix('0')+'\n'
#печатаем
#сложно описать ход мыслей - в муках рождалось)) кривовато, наверное, получилось, но вроде работает
for k, v in tab2.items():
    print(k)
    print(v)
````
